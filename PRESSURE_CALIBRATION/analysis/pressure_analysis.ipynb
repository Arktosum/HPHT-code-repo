```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
from datetime import datetime
```

```python
df = pd.read_csv("./pressure_resistance-FORWARD-1741345635.1614451.csv")

df.head()
```

```python
plt.figure(figsize=(16,9))

plt.semilogy(df['Load Pressure (bar)'], df['Resistance (ohm)']/0.011428635124729749, label='')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis (log scale)')
plt.title('Semilogy Plot Example')
plt.legend()

# Display the plot
plt.grid(True, which="both", ls="--")
plt.show()
```
