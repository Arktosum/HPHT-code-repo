```python
new_var = df['BATCHNO']
new_var.unique()
```

```python
calib = df[df['BATCHNO'] == 'CHN_PYRO_HT TRIAL 08_250724']
calib = calib.sort_values(by='TIME1')

calib = calib[calib['HYDPRESSURE1'] > 10]
calib.to_csv('calib_data.csv', index=False)
calib.head()
```

```python
# servo pressure set
# position ack

pressure_set = calib['HYDPRESSURE1'] + calib['HYDPRESSURE2'] + calib['HYDPRESSURE3'] + calib['HYDPRESSURE4'] + calib['HYDPRESSURE5'] + calib['HYDPRESSURE6']
pressure_set = pressure_set/6
import matplotlib.pyplot as plt
import numpy as np



# Initial Butting Position
# BUTTING_POSITIONS = [68.34	,2	,2,	68.85,	16.01,	3.05]
BUTTING_POSITIONS = [0	,0	,0,	0,	0,	0]
position=[]
for i in range(1,6+1):
    pos = calib[f'ACKPOSITIONCH{i}'] - BUTTING_POSITIONS[i-1]
    position.append(pos)
    

average_movement_inside_cube = np.array(position).mean(axis=0)

plt.figure(figsize=(16,20))
plt.grid(True)
plt.scatter(calib['HYDPRESSURE1'],average_movement_inside_cube,color='black',label='Top')
plt.scatter(calib['HYDPRESSURE2'],average_movement_inside_cube,color='red',label='Bottom')
plt.scatter(calib['HYDPRESSURE3'],average_movement_inside_cube,color='blue',label='Left')
plt.scatter(calib['HYDPRESSURE4'],average_movement_inside_cube,color='green',label='Right')
plt.scatter(calib['HYDPRESSURE5'],average_movement_inside_cube,color='yellow',label='Front')
plt.scatter(calib['HYDPRESSURE6'],average_movement_inside_cube,color='pink',label='Rear')
plt.legend(loc='upper left')

plt.show()
```
